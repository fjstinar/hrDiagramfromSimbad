```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('SIMBAD_NGC___104_csv.csv')
df = df.iloc[1:]
df.head()
```

```python
df.columns
```

```python
magB = df['B ( mag )']
magV = df['V ( mag )']
magB = magB[:-1]
magV = magV[:-1]
magB = list(map(float, magB))
magV = list(map(float, magV))
magB = np.array(magB)
magV = np.array(magV)
```

```python
plt.plot((magB - magV), magV, 'bo')
plt.xlabel('Temp')
plt.ylabel('Luminosity')
plt.title('H-R Diagram')
```
